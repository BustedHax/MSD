```python
#Problem 1: passive brownian motion in 1-D

import numpy as np
import matplotlib.pyplot as plt

N     = 2000
dt    = 0.01
D     = 3
runs  = 500

r     = np.zeros(N)
t     = r.copy()
msd   = r.copy()
m     = r.copy()

for i in range(runs):
    rng   = np.random.normal(0, 1, N)
    for i in range(1,N):
        r[i]    = r[i-1]+rng[i]*(dt*2*D)**0.5
        t[i]    = i
        msd[i]  = msd[i] + (r[i])**2
        m[i]    = m[i]   + r[i]

msd = msd/runs
m   = m/runs
        




fig, (ax1, ax2) = plt.subplots(2)
ax1.plot(t, r)
ax1.set(xlabel='t', ylabel='<r(t)>')
ax2.loglog(t, msd)
ax2.set(xlabel='t', ylabel='<r(t)²>')

slope0 = msd[-1]/(dt*N)
print("2*D = ", end='')
print(2*D)
print("MSD Slope = ", end='')
print(slope0)
```
